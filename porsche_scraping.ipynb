```python
import requests
from bs4 import BeautifulSoup as bs
import time
```

```python
time.sleep(2)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.5414.121 Safari/537.36 AOLShield/109.0.5414.12"
}

years = []
models = []
bodies = []
mileages = []
powers = []
cylinder_capacities = []
steerings = []
transmissions = []
drives = []
fuels = []
interior_materials = []
interior_colors = []
exterior_colors = []
conditions = []
new_used = []
ready_to_drive = []
locations = []

for x in range(1, 94):
    
    r = requests.get(f"https://www.elferspot.com/en/search/page/{x}/", headers=headers)
    soup = bs(r.content, "lxml")
    
    productlist = soup.find_all("a", class_="content-teaser")
    
    for link in productlist:
        href = link.get('href')
        
        car_url = f"https://www.elferspot.com{href}" if href.startswith('/') else href
        
        car_r = requests.get(car_url, headers=headers)
        car_soup = bs(car_r.content, "lxml")
        
        tables = car_soup.find_all("table", class_="fahrzeugdaten")
        
        for table in tables:
            rows = table.find_all("tr")
            
            for row in rows:
                label = row.find("td", class_="label")
                content = row.find("td", class_="content")
                
                if label and content:
                    label_text = label.get_text(strip=True)
                    link = content.find("a")
                    if link:
                        content_text = link.get_text(strip=True)
                    else:
                        content_text = content.get_text(strip=True)

                    if label_text == "Year of construction:":
                        years.append(content_text)
                    elif label_text == "Model:":
                        models.append(content_text)
                    elif label_text == "Body:":
                        bodies.append(content_text)
                    elif label_text == "Mileage:":
                        mileages.append(content_text)
                    elif label_text == "Power:":
                        powers.append(content_text)
                    elif label_text == "Cylinder capacity:":
                        cylinder_capacities.append(content_text)
                    elif label_text == "Steering:":
                        steerings.append(content_text)
                    elif label_text == "Transmission:":
                        transmissions.append(content_text)
                    elif label_text == "Drive:":
                        drives.append(content_text)
                    elif label_text == "Fuel:":
                        fuels.append(content_text)
                    elif label_text == "Interior material:":
                        interior_materials.append(content_text)
                    elif label_text == "Interior color:":
                        interior_colors.append(content_text)
                    elif label_text == "Exterior color:":
                        exterior_colors.append(content_text)
                    elif label_text == "Condition:":
                        conditions.append(content_text)
                    elif label_text == "New / used:":
                        new_used.append(content_text)
                    elif label_text == "Ready to drive:":
                        ready_to_drive.append(content_text)
                    elif label_text == "Car location:":
                        locations.append(content_text)

print("Years:", years)
print("Models:", models)
print("Bodies:", bodies)
print("Mileages:", mileages)
print("Powers:", powers)
print("Cylinder Capacities:", cylinder_capacities)
print("Steerings:", steerings)
print("Transmissions:", transmissions)
print("Drives:", drives)
print("Fuels:", fuels)
print("Interior Materials:", interior_materials)
print("Interior Colors:", interior_colors)
print("Exterior Colors:", exterior_colors)
print("Conditions:", conditions)
print("New/Used:", new_used)
print("Ready to Drive:", ready_to_drive)
print("Locations:", locations)
```
